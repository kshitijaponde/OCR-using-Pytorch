```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path1="pexels-kat-wilcox-923681.jpg"
img_path2="pexels-mwabonje-ringa-1806900.jpg"
```

```python
reader=easyocr.Reader(['en'], gpu=False)
result=reader.readtext(img_path2)
result
```

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
text=result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img=cv2.imread(img_path2)
img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),20)
img=cv2.putText(img, text, top_left, font, 5, (255,255,255), 25, cv2.LINE_AA)

plt.imshow(img)
plt.show()
```
